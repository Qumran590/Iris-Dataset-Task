```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('iris.csv')
```

```python
## Dataset structure
## We'll use 'shape','column',and 'info()' to understand the structure of the dataset.

#Number of rows and columns
print("Shape of the dataset:",df.shape)

#Column names
print("\nColumns:")
print(df.columns)

#Data types
print("\nData types:")
print(df.dtypes)

#First few rows
print("\nFirst 5 rows:")
print(df.head())

#Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
## Data cleaning
# We check for any missing values or incorrect data types.
```

```python
# check for missing values
if 'Id' in df.columns:
    df = df.drop('Id',axis = 1)
```

```python
## Data Visualization
# Now we explore the data using visualizations to understand pattrens and relationships.
```

```python
# Histogram for each numeric column
df.hist(figsize = (10,8))
plt.suptitle("Histograms of Iris Features",fontsize =16)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(data = df)
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
#Visualize Pairwise Realatinship
sns.pairplot(df,hue = "Species")
plt.suptitle("Pairplot of Iris Dataset",y = 1.02)
plt.show()
```

```python
print(df.dtypes)

# Convert only numeric columns
df_numeric = df.select_dtypes(include=['number'])

#Calculate the correlation matrix
corr=df_numeric.corr()

#Now plot heatmap
plt.figure(figsize =(8, 6))
sns.heatmap(corr,annot=True, cmap ='Blues')
plt.title("Correlation Heatmap")
plt.show()
```
